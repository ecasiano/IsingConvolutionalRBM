```python
import numpy as np
import matplotlib.pyplot as plt
import scipy 
from scipy import optimize
```

```python
T = 2.27
```

```python
colors = ['salmon','dodgerblue','mediumseagreen']
colors = ['#FFA0A0','#1e90ff','#3cb371']
colors_light = ['#FFEDED','#6BDDFF','#89FFBE']
```

```python
# T = 2.4
data_MC = np.loadtxt('./data_nnn/L_many_T_'+str(T)+'_J1_-1.0_J2_0.0_seed_1968_autocorr.dat')[:]
data_CRBM = np.loadtxt('./data_nnn/L_many_T_'+str(T)+'_J1_-1.0_J2_0.0_kernelDims_2-2_autocorr_no_regu.dat')[:]
data_reguCRBM = np.loadtxt('./data_nnn/L_many_T_'+str(T)+'_J1_-1.0_J2_0.0_kernelDims_2-2_autocorr_regu.dat')[:]

L_MC = data_MC[:,0]
tau_MC = data_MC[:,1]

L_CRBM = data_CRBM[:,0]
tau_CRBM = data_CRBM[:,1]

L_reguCRBM = data_reguCRBM[:,0]
tau_reguCRBM = data_reguCRBM[:,1]

# tau_err = data[:,2]
```

```python
# Plot
fig,ax = plt.subplots()

ax.scatter(L_CRBM,tau_CRBM,color=colors[1],fc=colors_light[1],label='CRBM')
ax.scatter(L_reguCRBM,tau_reguCRBM,color=colors[2],fc=colors_light[2],label='reguCRBM')
ax.scatter(L_MC,tau_MC,color=colors[0],fc=colors_light[0],label='MC')
ax.plot(L_CRBM,tau_CRBM,color=colors[1])
ax.plot(L_reguCRBM,tau_reguCRBM,color=colors[2])
ax.plot(L_MC,tau_MC,color=colors[0])
# ax.set_yscale("log")
# ax.set_xscale("log")
ax.set_ylabel(r"$\tau$ (energy)")
ax.set_xlabel(r"$L$")
ax.xaxis.set_tick_params(direction='in', which='both')
ax.yaxis.set_tick_params(direction='in', which='both')
ax.annotate(r"T=%.2f"%T,(0.7,0.4),xycoords='axes fraction')

ax.legend(frameon=True,loc='upper left')

plt.savefig('./figures/L_many_T_'+str(T)+'_J1_-1.0_J2_0.0_seed_1968_kernelDims_2-2_autocorr_E.pdf',dpi=300)
```
